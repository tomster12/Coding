```python
from PIL import Image

im = Image.open("justtop.png")
pix = im.load()
width, height = im.size
print("Size: ", width, height)
```

```python
im.resize((width * 4, height * 4))
```

```python
# Extract rows from image
rows = []
for y in range(height):
  rowLength = 2 * y + 6
  rowStart = 103 - y

  rows.append([])
  for x in range(rowLength):
    rows[y].append(1 - pix[rowStart + x, y])

# Get lengths of each row
lengths = [ len(row) for row in rows ]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Show as left aligned
imAligned = np.zeros((height, width), 'i4')
for y, row in enumerate(rows):
  for x, p in enumerate(row):
    imAligned[y, x] = p

plt.figure(0, (22, 22 * height / width))
plt.imshow(imAligned, cmap = "gray", interpolation = "nearest")
None
```
