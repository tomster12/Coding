```python
def map_c_to_v(c):
    return string.ascii_uppercase.index(c)

def map_v_to_c(c):
    return string.ascii_uppercase[c]

def safe_mod(v, l):
    while v < 0:
      v += l
    return v % l

alphabet = "OMAZWPHCEISRYVKGBFUNTXQJDL"
print(alphabet)
print("")

# Try each 2 <= d < 26 multiplier for decimation
outputs = []
for d in range(3, 26):

  # Try each rotation of the candidate alphabet
  for rot in range(len(alphabet)):
    rot_alphabet = (alphabet[-rot:] if rot > 0 else "") + alphabet[:len(alphabet) - rot]
    rot_alphabet_v = [ map_c_to_v(c) for c in rot_alphabet ]
    # print(rot_alphabet)

    # Reverse decimation to get new_alphabet_v and new_alphabet_v_inv
    new_alphabet_v_inv = { }
    for i, v in enumerate(rot_alphabet_v):
      new_alphabet_v_inv[safe_mod(d * i, 26)] = v

    # Print output if possible
    try:
      new_alphabet_v = { }
      for i in range(26):
        new_alphabet_v[new_alphabet_v_inv[i]] = i
      output = "".join([ map_v_to_c(new_alphabet_v[i]) for i in range(26) ])
      final_output = str(d) + "," + str(rot) + " : " + output
      outputs.append(final_output + "\n")
    except: pass

# Print outputs to file so can see
with open("test.txt", "w") as f:
  f.writelines(outputs)
```

```python
def encode_progressive(alphabet, plaintext, offset, increase):
    ciphertext = ""
    for l in plaintext:
        index = alphabet.index(l)
        next_index = (index + offset) % len(alphabet)
        ciphertext += alphabet[next_index]
        offset += increase
    return ciphertext
```

```python
alphabet = "ABCDEFGHIJ"
plaintext1 = "FIAAGHBCJFFBA"
plaintext2 = "HBCDDAAGHBCJFFBA"
plaintext3 = "HBDAAGHBCJFFBABCDFG"
ciphertext1 = encode_progressive(alphabet, plaintext1, 0, 2)
ciphertext2 = encode_progressive(alphabet, plaintext2, 0, 2)
ciphertext3 = encode_progressive(alphabet, plaintext3, 0, 2)
print(f'CT1: {ciphertext1}')
print(f'CT2: {ciphertext2}')
print(f'CT3: {ciphertext3}')

isomorph_12 = get_isomorphs(ciphertext1, ciphertext2)[0]
isomorph_12_1 = ciphertext1[isomorph_12[0]:isomorph_12[1]]
isomorph_12_2 = ciphertext2[isomorph_12[4]:isomorph_12[5]]
isomorph_23 = get_isomorphs(ciphertext2, ciphertext3)[0]
isomorph_23_1 = ciphertext2[isomorph_23[0]:isomorph_23[1]]
isomorph_23_2 = ciphertext3[isomorph_23[4]:isomorph_23[5]]
isomorph_13 = get_isomorphs(ciphertext1, ciphertext3)[0]
isomorph_13_1 = ciphertext1[isomorph_13[0]:isomorph_13[1]]
isomorph_13_2 = ciphertext3[isomorph_13[4]:isomorph_13[5]]
print(isomorph_12)
print(isomorph_23)
print(isomorph_13)
print(f'Isomorph 12 1: {isomorph_12_1}')
print(f'Isomorph 12 2: {isomorph_12_2}')
print(f'Isomorph 23 1: {isomorph_23_1}')
print(f'Isomorph 23 2: {isomorph_23_2}')
print(f'Isomorph 13 1: {isomorph_13_1}')
print(f'Isomorph 13 2: {isomorph_13_2}')

chains12 = get_chains(isomorph_12_1, isomorph_12_2)
chains23 = get_chains(isomorph_23_1, isomorph_23_2)
chains13 = get_chains(isomorph_13_1, isomorph_13_2)
print(chains12)
print(chains23)
print(chains13)

print(len(alphabet))
```
