```python
from locallib import analysis, eyeutil
import matplotlib.pyplot as plt
import numpy as np
import string
```

```python
analysis.plot_im(
    im=analysis.conv_msgs_to_im(eyeutil.msgs_eye_tri),
    title="Plain Messages (CIPHERTEXT_NUM)",
    figsize=(28,2))

analysis.plot_im(
    im=analysis.conv_msgs_to_im(eyeutil.msgs_eye_tri),
    ascii=True,
    title="Plain Messages (CIPHERTEXT_ASCII)",
    figsize=(28,2))

shared_im = analysis.calc_shared(msgs=eyeutil.msgs_eye_tri, to_zero=True)
analysis.plot_im(
    im=shared_im,
    labels=eyeutil.msgs_eye_tri,
    title="Shared Sections",
    to_dull=True,
    under_value=1,
    figsize=(28,2))

gap_im_no_ends = analysis.calc_gaps(msgs=eyeutil.msgs_eye_tri, limit=16, use_msg_value=False, to_zero=True)
analysis.plot_im(
    im=gap_im_no_ends,
    to_label=True,
    labels=eyeutil.msgs_eye_tri,
    title="Gaps (start only, msg value)",
    to_dull=True,
    figsize=(28,2))

gap_im = analysis.calc_gaps(msgs=eyeutil.msgs_eye_tri, limit=16, include_end=True, use_msg_value=False, to_zero=True)
analysis.plot_im(
    im=gap_im,
    to_label=True,
    labels=eyeutil.msgs_eye_tri,
    title="Gaps (start+end, msg value)",
    to_dull=True,
    figsize=(28,2))
analysis.plot_im(
    im=gap_im,
    to_label=True,
    title="Gaps (start+end, gap size)",
    to_dull=True,
    figsize=(28,2))

im_repeats = analysis.calc_repeats(eyeutil.msgs_eye_tri)
analysis.plot_im(
    im=im_repeats,
    title="Eyes Repeats (REPEAT_COUNT)")

msgs_random, _ = eyeutil.generate_random_msgs_tri()
im_repeats = analysis.calc_repeats(msgs_random)
analysis.plot_im(
    im=im_repeats,
    title="Repeats Random (REPEAT_COUNT)")

plt.show()
```

```python
msgs19 = eyeutil.msgs_eye_tri[0:9]
isos_im = analysis.conv_isomorphs_to_img(msgs19, analysis.calc_isomorphs(msgs19), 0)
analysis.plot_im(
    im=isos_im,
    to_label=True,
    to_dull=True,
    labels=msgs19,
    title="(1-9) Isomorphs",
    figsize=(28,2))

msgs13 = eyeutil.msgs_eye_tri[0:3]
isos_im = analysis.conv_isomorphs_to_img(msgs13, analysis.calc_isomorphs(msgs13), 0)
analysis.plot_im(
    im=isos_im,
    to_label=True,
    to_dull=True,
    labels=msgs13,
    title="(1-3) Isomorphs",
    figsize=(28,2))

msgs46 = eyeutil.msgs_eye_tri[3:6]
isos_im = analysis.conv_isomorphs_to_img(msgs46, analysis.calc_isomorphs(msgs46), 0)
analysis.plot_im(
    im=isos_im,
    to_label=True,
    to_dull=True,
    labels=msgs46,
    title="(4-6) Isomorphs",
    figsize=(28,2))

msgs79 = eyeutil.msgs_eye_tri[6:9]
isos_im = analysis.conv_isomorphs_to_img(msgs79, analysis.calc_isomorphs(msgs79), 0)
analysis.plot_im(
    im=isos_im,
    to_label=True,
    to_dull=True,
    labels=msgs79,
    title="(7-9) Isomorphs",
    figsize=(28,2))

plt.show()
```
